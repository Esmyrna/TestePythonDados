```python
# Você trabalha em uma empresa que vende bermudas e essa empresa tem 5 lojas e quer aumentar suas vendas;
# A única informação que se tem sobre isso é uma base de dados.
```

Passo 1: Trazer sua base de dados para Python e verificar o que há dentro dela.

```python
import pandas as pd

tabela = pd.read_excel("Vendas.xlsx")
display(tabela)
```

Passo 2: Pegar um panorama geral da base de dados





```python
# ver como está o faturamento da empresa, o quanto ela vendeu. Como? somando o valor final da empresa
```

```python
faturamento = tabela["Valor Final"].sum()
display(faturamento)
```

```python
# Será que tem uma loja vendendo mais que a outra? Como se chegou nesse resultado? 
```

Passo 3 : Começar sua análise

```python
# Faturamento por loja:

faturamentoPorLoja = tabela[["ID Loja" , "Valor Final"]]
display(faturamentoPorLoja)

```

Passo 4: Agrupar as lojas com nomes iguais

```python
faturamentoPorLoja = tabela[["ID Loja" , "Valor Final"]].groupby("ID Loja").sum()
display(faturamentoPorLoja)
```

Passo 5: Ver o que realmente está fazendo a loja Iguatemi Campinas vender mais!

```python
# se for o produto, iremos agrupar loja, produto e valor final
faturamentoPorLoja = tabela[["ID Loja" ,"Produto" ,"Valor Final"]].groupby([ "ID Loja" , "Produto" ]).sum()
display(faturamentoPorLoja)
```

Conclui-se que a Bermuda Lisa é a que mais vendeu, então deve-se aplicar esse produto nas lojas restantes a fim de faturar mais.
